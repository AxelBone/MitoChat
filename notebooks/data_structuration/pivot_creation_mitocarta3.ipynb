### Fonctions utilitaires

```python
import unicodedata

def normalize_text(s):
    """Supprime espaces et accents, normalise le texte"""
    if pd.isna(s):
        return None
    s = str(s).strip()
    s = unicodedata.normalize('NFKD', s)
    return s

import re

def split_field_to_list(s):
    """Transforme un champ séparé par ; , | / en liste"""
    if pd.isna(s):
        return []
    parts = re.split(r'[;,|/]+', str(s))
    return [p.strip() for p in parts if p.strip()]
```

```python
# Détection de la structure et du nettoyage à appliquer.
mitocarta['Synonyms_list'] = mitocarta['Synonyms'].apply(split_field_to_list)
mitocarta['MitoCarta3_Pathways_list'] = mitocarta['MitoCarta3.0_MitoPathways'].apply(split_field_to_list)
mitocarta['MitoCarta3_SubMitoLoc_list'] = mitocarta['MitoCarta3.0_SubMitoLocalization'].apply(split_field_to_list)

print(mitocarta["Synonyms_list"].head())
```

```python
pivot_records = {}

for _, row in mitocarta.iterrows():
    hgnc_id = row['HumanGeneID']
    if pd.isna(hgnc_id):
        continue  # sauter les lignes sans HumanGeneID

    record = {
        "_key": hgnc_id,
        "human_gene_id": hgnc_id,
        "symbol": normalize_text(row['Symbol']),
        "synonyms": split_field_to_list(row['Synonyms']),
        "description": normalize_text(row['Description']),
        "ensembl_id": normalize_text(row['EnsemblGeneID_mapping_version_20200130']),
        "uniprot_ids": split_field_to_list(row['UniProt']),
        "mouse_ortholog_gene_id": normalize_text(row['MouseOrthologGeneID']),
        
        "mitocarta3": {
            "list_status": row['MitoCarta3.0_List'] == "yes",
            "evidence": split_field_to_list(row['MitoCarta3.0_Evidence']),
            "sub_mito_localization": split_field_to_list(row['MitoCarta3.0_SubMitoLocalization']),
            "mito_pathways": split_field_to_list(row['MitoCarta3.0_MitoPathways'])
        },

        "mitocarta2": {
            "score": row['MitoCarta2.0_Score'],
            "fdr": row['MitoCarta2.0_FDR'],
            "list_status": row['MitoCarta2.0_List'] == "yes",
            "evidence": split_field_to_list(row['MitoCarta2.0_Evidence'])
        },

        "training_dataset": row['TrainingDataset'],
        "protein_length": row['ProteinLength'],
        "scores": {
            "targetp": row['TargetP_Score'],
            "mitodomain": row['MitoDomain_Score'],
            "coexpression": row['CoexpressionGnfN50_Score'],
            "pgc_induction": row['PGC_Induction_Score'],
            "yeast_homolog": row['YeastMitoHomolog_Score'],
            "rickettsia_homolog": row['RickettsiaHomolog_Score']
        },

        "msms": {
            "score": row['MSMS_Score'],
            "num_tissues": row['MSMS_NUM_TISSUES'],
            "num_peptides_unique": row['MSMS_NUM_PEPTIDES_UNIQUE'],
            "num_spectra": row['MSMS_NUM_SPECTRA'],
            "total_intensity": row['MSMS_TOTAL_INTENSITY'],
            "percent_coverage": row['MSMS_PERCENT_COVERAGE'],
            "by_tissue_log10": {
                "cerebrum": row['cerebrum_total_peak_intensity_log10'],
                "cerebellum": row['cerebellum_total_peak_intensity_log10'],
                "brainstem": row['brainstem_total_peak_intensity_log10'],
                "spinalcord": row['spinalcord_total_peak_intensity_log10'],
                "kidney": row['kidney_total_peak_intensity_log10'],
                "liver": row['liver_total_peak_intensity_log10'],
                "heart": row['heart_total_peak_intensity_log10'],
                "skeletalmuscle": row['skeletalmuscle_total_peak_intensity_log10'],
                "adipose": row['adipose_total_peak_intensity_log10'],
                "smallintestine": row['smallintestine_total_peak_intensity_log10'],
                "largeintestine": row['largeintestine_total_peak_intensity_log10'],
                "stomach": row['stomach_total_peak_intensity_log10'],
                "placenta": row['placenta_total_peak_intensity_log10'],
                "testis": row['testis_total_peak_intensity_log10']
            }
        },

        "genome_coords_hg19": {
            "chromosome": row['hg19_Chromosome'],
            "start": row['hg19_Start'],
            "stop": row['hg19_Stop']
        },

        "hpa_location_2020": row['HPA_Main_Location_2020 (Reliability)']
    }

    pivot_records[hgnc_id] = record
```

```python
import json, os, datetime

outdir = "../data/pivot_output"
os.makedirs(outdir, exist_ok=True)

# JSONL (une ligne par gene)
jsonl_path = os.path.join(outdir, "mitocarta_pivot.jsonl")
with open(jsonl_path, 'w', encoding='utf8') as f:
    for rec in pivot_records.values():
        rec['_created_at'] = datetime.datetime.utcnow().isoformat() + 'Z'
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")

print("Pivot JSONL sauvegardé :", jsonl_path)

# JSON complet
json_path = os.path.join(outdir, "mitocarta_pivot_full.json")
with open(json_path, 'w', encoding='utf8') as f:
    json.dump(pivot_records, f, ensure_ascii=False, indent=2)

print("Pivot JSON complet sauvegardé :", json_path)
```
